```python
import plotly.graph_objs as go
import numpy as np
import plotly.offline as offline
import pandas as pd
```

```python
hmda_data = pd.read_csv('../data/hmda_short.csv')

hmda_data
```

```python
trace0 = go.Scatter(

    x = hmda_data['census_tract'],
    y = hmda_data['loan_amount'],
    name = 'loan amount by lei',
    line = dict(
                    color = ('rgb(0, 260, 74)'),
                    width = 4,
                    dash = 'solid'
                )
)
```

```python
data = [trace0]

layout = dict(
                title = 'HMDA POC Data Engineering',
                xaxis = dict(title = 'Census Tract'),
                yaxis = dict(title = 'Loan Amount'),
             )
```

```python
fig = dict(data = data, layout = layout)

offline.iplot(fig)
```

```python
data = [go.Bar(
                y = hmda_data['census_tract'],
                x = hmda_data['loan_amount'],
                )]
offline.iplot(data)
```
